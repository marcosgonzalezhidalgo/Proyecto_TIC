```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configuración de estilos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Ruta a los resultados de VM
results_path = os.path.abspath(os.path.join("..", "results"))
vm_csv = os.path.join(results_path, "vm_scenarios_results.csv")

# Cargar datos
df_vm = pd.read_csv(vm_csv)
df_vm['timestamp'] = pd.to_datetime(df_vm['timestamp'], format='%Y%m%d_%H%M%S')

# Ordenar por tiempoe
df_vm.sort_values('timestamp', inplace=True)

# Mostrar las últimas filas
df_vm.tail()
```

```python
sns.barplot(data=df_vm, x="test_type", y="cpu_percent", hue="source")

plt.title("Uso promedio de CPU (%) por tipo de carga")
plt.xlabel("Tipo de prueba")
plt.ylabel("CPU (%)")
plt.legend(title="Fuente")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df_vm, x="test_type", y="memory_percent", hue="source")

plt.title("Uso promedio de Memoria (%) por tipo de carga")
plt.xlabel("Tipo de prueba")
plt.ylabel("Memoria (%)")
plt.legend(title="Fuente")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df_vm, x="test_type", y="execution_time_sec", hue="source")

plt.title("Tiempo de ejecución por tipo de carga")
plt.xlabel("Tipo de prueba")
plt.ylabel("Segundos")
plt.legend(title="Fuente")
plt.tight_layout()
plt.show()
```
