```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Rutas de los archivos
results_path = os.path.abspath(os.path.join("..", "results"))
vm_csv = os.path.join(results_path, "vm_benchmark_results.csv")
# docker_csv = os.path.join(results_path, "docker_benchmark_results.csv")  # Si luego lo tienes

# Cargar datos
df_vm = pd.read_csv(vm_csv)
df_vm['source'] = 'VM'

# df_docker = pd.read_csv(docker_csv)
# df_docker['source'] = 'Docker'

# df_all = pd.concat([df_vm, df_docker], ignore_index=True)
df_all = df_vm.copy()

df_all.tail()
```

```python
plt.figure(figsize=(10,5))
for source in df_all['source'].unique():
    subset = df_all[df_all['source'] == source]
    plt.plot(subset['timestamp'], subset['cpu_percent'], marker='o', label=source)

plt.title("Uso promedio de CPU (%)")
plt.xlabel("Tiempo")
plt.ylabel("CPU %")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
for source in df_all['source'].unique():
    subset = df_all[df_all['source'] == source]
    plt.plot(subset['timestamp'], subset['memory_percent'], marker='s', label=source)

plt.title("Uso promedio de Memoria (%)")
plt.xlabel("Tiempo")
plt.ylabel("Memoria %")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
